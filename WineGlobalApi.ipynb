```python
import urllib
import json

# Import the API key.
from config import wine_api_key
```

## Red Wine for Usa:

```python
headers = {
    'accept': 'application/json',
    'authorization': wine_api_key
    }
req = urllib.request.Request('https://api.globalwinescore.com/globalwinescores/latest/?color=red&limit=10000&country=Usa', data=None, headers=headers, origin_req_host=None, unverifiable=False, method=None)
```

```python
res = urllib.request.urlopen(req).read().decode()
obj = json.loads(res)
```

```python
obj
```

```python
#Create a dataframe for red_wine
import pandas as pd
red_wine_df = pd.DataFrame.from_dict(obj["results"], orient='columns')
red_wine_df
```
